```python
# Refer to the attached CSV file. Divide a csv 
# in ‘n’ chunks and create ‘n’ new csv files.
```

```python
import pandas as pd
```

```python
# Create DataFrame from CSV data

df = pd.read_csv("~/Documents/practice_programs/python_programs/assignment/Zestl-Assignments/sample.csv")
df.head()
```

```python
df.info()
```

```python
# Chunk the DataFrame based on the Item Type

item_types = df["Item Type"].unique()
print(item_types)
print(len(item_types))
```

```python
# Create a group of common Item Type as a pandas groupby object

item_types_df = df.groupby("Item Type")
```

```python
# Iterate through the grouped Item Type pandas groupby object
# and generate the chunked data as CSV files

i = 0
while i < len(item_types_df):
    new_df = pd.DataFrame(data=list(item_types_df)[i][1])
    new_df.to_csv("csv_outputs/item_types/item_type-"+ item_types[i] +".csv", index=False)
    i += 1
```

```python
# SAMPLE OUTPUT OF THE CHUNKED DATA - Office Supplies (Item Type)

pd.read_csv("csv_outputs/item_types/item_type-Office Supplies.csv")
```

```python
# Chunk the DataFrame based on the Region

regions = df["Region"].unique()
print(regions)
print(len(regions))
```

```python
# Create a group of common Region as a pandas groupby object

region_df = df.groupby('Region')
```

```python
# Iterate through the grouped Region pandas groupby object
# and generate the chunked data as CSV files

i = 0
while i < len(region_df):
    new_df = pd.DataFrame(data=list(region_df)[i][1])
    new_df.to_csv("csv_outputs/regions/region-"+ regions[i] +".csv", index=False)
    i += 1
```

```python
# SAMPLE OUTPUT OF THE CHUNKED DATA - Australia and Oceania (Region)

pd.read_csv("csv_outputs/regions/region-Australia and Oceania.csv")
```
